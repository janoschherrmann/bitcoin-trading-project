```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pot

df=pd.read_csv('abc.csv')


df #before changing the base index
```

```python
df = df.set_index(pd.DatetimeIndex(df['Date'].values)) #setting index as date as we make change in order to the date
df #after chaging the base index
```

```python
# Get the 26-day EMA of the closing price
k = df['Close'].ewm(span=12, adjust=False, min_periods=12).mean()
# Get the 12-day EMA of the closing price
d = df['Close'].ewm(span=26, adjust=False, min_periods=26).mean()
# Subtract the 26-day EMA from the 12-Day EMA to get the MACD
macd = k - d
# Get the 9-Day EMA of the MACD for the Trigger line
macd_s = macd.ewm(span=9, adjust=False, min_periods=9).mean()
# Calculate the difference between the MACD - Trigger for the Convergence/Divergence value
macd_h = macd - macd_s
# Add all of our new values for the MACD to the dataframe
df['macd'] = df.index.map(macd)
df['macd_h'] = df.index.map(macd_h)
df['macd_s'] = df.index.map(macd_s)
# View our data
pd.set_option("display.max_columns", None)
df

# add columns to table
column_list =[ 'macd', 'macd_h', 'Close']
#ploting the given data in chart
df[column_list].plot(figsize=(20,10)) #determining figure size


pot.title('MACD Bitcoin Trading Strategy') #title for the Chart
pot.ylabel('USD Price in $') # labeling the y axis of the given Chart
pot.xlabel('Date in years') # labeling the x axis of the given Chart
pot.yticks(rotation = 15) #rotation of price by 15 degree 
pot.show() #Display the  Graph

new_df = df[45-1:]
new_df
```

```python
def get_signal(data):

  buy_signal = []
  sell_signal = []

  for i in range(len(data['Close'])):
    if i == 0:
      pass
    if data['macd'][i - 1] < data['macd_s'][i - 1] and data['macd'][i] >= data['macd_s'][i]: # then you should sell
      buy_signal.append(data['Close'][i])
      sell_signal.append(np.nan)
    elif data['macd'][i - 1] > data['macd_s'][i - 1] and data['macd'][i] <= data['macd_s'][i]: # then buy
      buy_signal.append(np.nan)
      sell_signal.append(data['Close'][i])
    else:
      buy_signal.append(np.nan)
      sell_signal.append(np.nan)

  return buy_signal, sell_signal

# Create two new columns
new_df['Buy'] = get_signal(new_df)[0]
new_df['Sell'] = get_signal(new_df)[1]

# plot all the data

fig = pot.figure(figsize=(30,15)) #size of graph
#add plot
gray = fig.add_subplot(1,1,1)
#get iindex values of data frame
x_axis = new_df.index
#plot closing price and average
gray.plot(x_axis, new_df['Close'], color ='red', lw =3, label='Close Price', alpha= 0.1)
gray.plot(x_axis, new_df['macd'], color ='blue', lw =3, label='MACD', alpha= 0.1)
gray.scatter(x_axis, new_df['Buy'], color ='green', lw=3 , label='Buy', marker='^', alpha= 1) #adding the green buy arrow , when the algorith tell us to buy
gray.scatter(x_axis, new_df['Sell'], color ='red', lw=3 , label='Sell', marker='v', alpha= 1) #adding the red sell arrow , when the algorith tell us to sell

gray.set_title('MACD Bitcoin trading strategy')
gray.set_xlabel('Date')
gray.set_ylabel('USD Price in $')
pot.yticks(rotation = 15) #rotation of price by 15 degree 
gray.legend()  #showing the legend
pot.show() #displaying the graph
pd.set_option("display.max_rows", 100)
```

```python

buy = []
sell = []
open_pos = False

for i in range(len(new_df)):
  if(i == 0):
    pass
  if new_df['macd'][i - 1] < new_df['macd_s'][i - 1] and new_df['macd'][i] >= new_df['macd_s'][i] and open_pos == False:
    buy.append(i)
    open_pos = True
  elif new_df['macd'][i - 1] > new_df['macd_s'][i - 1] and new_df['macd'][i] <= new_df['macd_s'][i] and open_pos:
    sell.append(i)
    open_pos = False

```

```python
total = pd.concat([df.iloc[buy].Close, df.iloc[sell].Close], axis = 1)
total.columns = ['Buys', 'Sells']
total
```

```python
totalprofit = total.shift(-1).Sells - total.Buys
totalprofit
```

```python
relprofit = (total.shift(-1).Sells - total.Buys)/total.Buys
relprofit
```
